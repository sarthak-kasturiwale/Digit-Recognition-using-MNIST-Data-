# Dimensionality Reduction and Visualization

# Code to load MNIST Data set

```python
# MNIST data set reading
#Functions to read and visualize the data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#reading MNIST data from csv file
data = pd.read_csv('mnist_train.csv')

#printing first five rows of the data
print(data.head(5))

#describing what it exactly contains
data.describe()
```

```python
#save the labels in the variable lbl
label = data['label']

#drop the label feature from data and store it into wldata variable
no_lbl_data = data.drop('label',axis = 1)

#printing both the data and labels
print(label.head())
print(no_lbl_data.head())
```

```python
#printing shape of the data and label
print(label.shape)
print(no_lbl_data.shape)
```

```python
#display a certain data and its c/r number
plt.figure(figsize = (8,8))
index = 110

#reshape is used coz it is a 1-D column vector so to convert it into 2-D
#FutureWarning: Method .as_matrix will be removed in a future version. Use .values instead.
#We have used : grid_data = wldata.iloc[index].as_matrix().reshape(28,28)
grid_data = no_lbl_data.iloc[index].values.reshape(28,28)
plt.imshow(grid_data, cmap = 'gray', interpolation = 'none')
plt.show()
```

```python
print(label[index])
```

# Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

std_data = sc.fit_transform(no_lbl_data)
std_data.shape
```

# PCA using Scikit-Learn

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2, random_state = 0)

pca_data = pca.fit_transform(std_data)
print(type(pca_data))
print('')

# pca_reduced will contain the 2-d projects of simple data
print("Shape of Reduced Data = ", pca_data.shape)
```

```python
print(pca_data.T.shape)
print(label.shape)
```

```python
# As it is a Numpy array we have to use stacking for attaching the Labels
pca_data_with_label = np.vstack((pca_data.T, label)).T
```

```python
# creating a new data fram which help us in ploting the result data
pca_df = pd.DataFrame(data = pca_data_with_label, columns = ("1st_principal", "2nd_principal", "label"))
print(pca_df)
```

```python
var_explained = pca.explained_variance_ratio_
print(var_explained)
```

```python
import seaborn as sns

sns.FacetGrid(data = pca_df, hue = 'label', size = 6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
```

# t-SNE using Scikit-Learn

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, random_state = 0)
tsne_data = tsne.fit_transform(std_data)
```

```python
# pca_reduced will contain the 2-d projects of simple data
print("Shape of Reduced Data ", tsne_data.shape)
print('')
print('Shape of Labeled Data', label.shape)
print('')

# As it is a Numpy array we have to use stacking for attaching the Labels
tsne_data_with_label = np.vstack((tsne_data.T, label)).T
```

```python
# creating a new data fram which help us in ploting the result data
tsne_df = pd.DataFrame(data = tsne_data_with_label, columns = ("1st_principal", "2nd_principal", "label"))
print(tsne_df)
```

```python

sns.FacetGrid(data = tsne_df, hue = 'label', size = 6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
```
